## VOICED Database Capstone Research
The VOICED (Voice Disorder Database) is a comprehensive collection of voice samples designed to facilitate the study of voice disorders. The database is publicly available on PhysioNet and contains a significant number of voice samples from both pathological and healthy subjects.

### Database Composition
The VOICED database is a rich resource that includes a total of 208 voice samples. It encompasses recordings from 150 individuals with voice pathologies and 58 from healthy voices. This diverse collection allows for a broad analysis of voice characteristics across different health statuses.

### Demographic and Clinical Information
In addition to voice samples, the database provides detailed demographic and clinical information. This includes the gender, age, and pathology of the participants, as well as lifestyle factors such as smoking, alcohol, and coffee consumption. Occupational status is also documented. Such comprehensive data can help researchers identify patterns and correlations between these variables and voice disorders.

### Participant Criteria
The study that contributed to the VOICED database had specific inclusion criteria. Participants were required to be adults, aged between 18 and 70 years, and capable of following the study protocol. This age range ensures that the database covers a wide spectrum of adult voices.

### Recording Methodology
Voice samples in the VOICED database were captured using Vox4Health, an m-health system. This system is designed to record voice signals in real-time using the microphone of a mobile device, ensuring that the recordings are easily accessible and can be made in various settings.

### Technical Specifications
The technical quality of the voice recordings is maintained at a high standard. All samples in the database were recorded with a sampling rate of 8000 Hz and a resolution of 32-bit. These specifications ensure that the recordings are of sufficient quality for detailed analysis.

### Data Format
The VOICED database provides the voice recordings in two formats: the WFDB format and text format. The WFDB format is particularly useful for researchers familiar with PhysioNet's tools, as it allows for easy manipulation and analysis of the data.

### Recording Environment
To ensure consistency and quality of the voice samples, all recordings were made in a controlled environment. The room was kept quiet, with less than 30 dB of background noise, and had a humidity level between 30-40% . Such conditions help to minimize external factors that could affect the voice recordings.

## Conclusion:
The VOICED database is a valuable resource for researchers interested in voice disorders. It provides a wealth of high-quality voice samples along with extensive demographic and clinical information, all of which are essential for a thorough analysis of voice health and pathology.

```python
import warnings
warnings.filterwarnings('ignore')
import os
from tqdm import tqdm
from PIL import Image
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pyts.image import GramianAngularField
import cv2

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from sklearn.experimental import enable_halving_search_cv
from sklearn.model_selection import HalvingGridSearchCV, StratifiedKFold
from sklearn.svm import SVC

from sklearn.metrics import classification_report, ConfusionMatrixDisplay

source_folder = "/Users/harisx91/Documents/Intro_Computer_Science/Capstone_6500/VOICED_DATASET"
target_folder = "/Users/harisx91/Documents/Intro_Computer_Science/Capstone_6500/more_imgs"


# Create the target folders if they don't exist
os.makedirs(os.path.join(target_folder, 'healthy'), exist_ok=True)
os.makedirs(os.path.join(target_folder, 'pathological'), exist_ok=True)

# List all files in the source folder
file_list = os.listdir(source_folder)

for filename in tqdm(file_list, desc="Processing files"):
    if filename.endswith('.txt'):
        # Read the data from the file
        file_path = os.path.join(source_folder, filename)
        try:
            data = pd.read_csv(file_path, delimiter='\t')
            
            # Drop non-numeric columns if any
            data = data.select_dtypes(include=np.number)
            
            array = data.values
            array = np.transpose(array)
            
            gasf = GramianAngularField(method='summation', image_size=250)
            img1 = gasf.transform(array)
            
            gadf = GramianAngularField(method='difference', image_size=250)
            img2 = gadf.transform(array)
            
            # Concatenate the transformed images
            img = np.concatenate((img1, img2, np.zeros((1, 250, 250))), axis=0)
            
            # Save the produced image in the target folder as jpg
            target_path = os.path.join(target_folder, filename.replace('.txt', '.jpg'))
            
            # Convert numpy array to PIL Image and save as jpg
            img = (img * 255).astype(np.uint8)
            img = np.transpose(img, (1,2,0))
            pil_img = Image.fromarray(img)
            pil_img.save(target_path)
        except Exception as e:
            print(f"Error processing file {filename}: {e}")
```

```python
SZ = 20
bin_n = 32 # Number of bins
affine_flags = cv2.WARP_INVERSE_MAP|cv2.INTER_LINEAR
```

```python
def deskew(img):
   img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
   m = cv2.moments(img)
   if abs(m['mu02']) < 1e-2:
     return img.copy()
   skew = m['mu11']/m['mu02'] ##(The skewness is calculated using the formula `skew = m['mu11']/m['mu02']`)
   M = np.float32([[1, skew, -0.5*SZ*skew], [0, 1, 0]])
   img = cv2.warpAffine(img,M,(SZ, SZ),flags=affine_flags)
   return img
```

Python function called `deskew` which is correcting skewness in images using OpenCV. This function performs the following steps:

1. Converts the input image to grayscale (`cv2.cvtColor`: Converts the input image to grayscale).
2. Calculates the moments of the grayscale image (`cv2.moments`: Calculates the moments of the grayscale image).
3. Checks if the second central moment (`mu02`) is close to zero.
4. Calculates the skewness of the image using the moments (The skewness is calculated using the formula `skew = m['mu11']/m['mu02']`).
5. Constructs an affine transformation matrix to deskew the image (- An affine transformation matrix `M` is constructed using the calculated skewness).
6. Applies the affine transformation to the image using `cv2.warpAffine`(- `cv2.warpAffine` applies the affine transformation to the image).
7. The function returns the deskewed image.

```python
def hog(img):
   gx = cv2.Sobel(img, cv2.CV_32F, 1, 0)
   gy = cv2.Sobel(img, cv2.CV_32F, 0, 1)
   mag, ang = cv2.cartToPolar(gx, gy)
   bins = np.int32(bin_n*ang/(2*np.pi))
   bin_cells = bins[:10,:10], bins[10:,:10], bins[:10,10:], bins[10:,10:]
   mag_cells = mag[:10,:10], mag[10:,:10], mag[:10,10:], mag[10:,10:]
   hists = [np.bincount(b.ravel(), m.ravel(), bin_n) for b, m in zip(bin_cells, mag_cells)]
   hist = np.hstack(hists) # hist is a 64 bit vector
   return hist
```

```python
hogged = hog(deskewed)
plt.plot(hogged)
plt.show()
```

Define a function called `hog` which computes the Histogram of Oriented Gradients (HOG) for a given image. After defining the `hog` function, the code computes the HOG feature vector for a deskewed image (`deskewed`) using this function. Then, it plots the HOG feature vector using `plt.plot(hogged)`.

### So, it computes the HOG feature vector for the deskewed image and plots it as a line plot. Each value in the plot represents a feature extracted from the image using the HOG algorithm.

```python
# Initialize variables
img_folder_1 = "/Users/harisx91/Documents/Intro_Computer_Science/Capstone_6500/more_imgs/healthy"
img_folder_2 = "/Users/harisx91/Documents/Intro_Computer_Science/Capstone_6500/more_imgs/pathological"
imgs = []
labels = []

# Process images in folder 1
for img_name in os.listdir(img_folder_1):
    img_path = os.path.join(img_folder_1, img_name)
    img = cv2.imread(img_path)
    if img is not None:
        img = deskew(img)
        img = hog(img)
        imgs.append(img)
        labels.append(0)

# Process images in folder 2
for img_name in os.listdir(img_folder_2):
    img_path = os.path.join(img_folder_2, img_name)
    img = cv2.imread(img_path)
    if img is not None:
        img = deskew(img)
        img = hog(img)
        imgs.append(img)
        labels.append(1)

# Convert the lists to NumPy arrays
imgs = np.array(imgs)
labels = np.array(labels)

# Print the shape of the arrays to verify
print("Images shape:", imgs.shape)
print("Labels shape:", labels.shape)
```

2. It iterates through the images in each folder, reads each image using OpenCV (`cv2.imread`), and performs deskewing and HOG feature extraction (`deskew` and `hog` functions) on each image.
3. It appends the processed image and its label (0 for healthy, 1 for pathological) to the `imgs` and `labels` lists.
4. After processing all images, it converts the lists to NumPy arrays.
5. Finally, it prints the shape of the arrays to verify the number of images processed and their dimensions.

```python
scaler = StandardScaler()
x = scaler.fit_transform(imgs)
```

**Standardization**:
   - `scaler = StandardScaler()`: Initializes a StandardScaler object.
   - `x = scaler.fit_transform(imgs)`: This code standardizes the dataset `imgs` using the `StandardScaler`. It subtracts the mean and scales each feature to unit variance.
   - `plt.plot(x, color='red')`: Plots the standardized dataset `x`.

```python
plt.plot(x, color='red')
plt.title('Standardized Dataset')
plt.show()
```

```python
##sm = SMOTE(random_state=42, sampling_strategy='minority')
##X_res, y_res = sm.fit_resample(x, labels)
sm = SMOTE(random_state=42, sampling_strategy='minority', k_neighbors=4)
X_res, y_res = sm.fit_resample(x, labels)
```

**Oversampling with SMOTE**:
   - `sm = SMOTE(random_state=42, sampling_strategy='minority')`: Initializes a SMOTE object with a specified random state and sampling strategy.
   - `X_res, y_res = sm.fit_resample(x, labels)`: Applies the SMOTE algorithm to the standardized dataset `x` and corresponding labels `labels`. SMOTE generates synthetic samples for the minority class (pathological images) to balance the class distribution.
   - `X_res` contains the resampled feature matrix, and `y_res` contains the corresponding resampled labels.

### Summary: Visualizes the original dataset, then standardizes it to have a mean of 0 and a standard deviation of 1. Finally, it applies the SMOTE algorithm to address class imbalance by oversampling the minority class (pathological images) and plots the standardized and resampled datasets for comparison.

```python
plt.plot(X_res, color='green')
plt.title('Augmented Dataset')
plt.show()
```

The augmented dataset obtained after applying the SMOTE algorithm.

- `plt.plot(X_res, color='green')`: This code plots the resampled feature matrix `X_res`, which represents the augmented dataset. Each row of `X_res` corresponds to a sample (image), and each column represents a feature extracted from that image (e.g., HOG features). The `color='green'` argument specifies the color of the plot.
  
- `plt.title('Augmented Dataset')`: Sets the title of the plot to 'Augmented Dataset'.

This plot visualizes the distribution of features in the augmented dataset, which now includes synthetic samples generated by the SMOTE algorithm to address class imbalance. By plotting this dataset, you can observe how the SMOTE algorithm has balanced the class distribution by oversampling the minority class (pathological images) to match the number of samples in the majority class (healthy images).

```python
clf = SVC(random_state=42)
cv = StratifiedKFold(shuffle=True, random_state=42)
param_grid = {'C': [0.1, 1, 10, 100, 1000], 
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001], 
              'kernel':['linear','rbf']}
search = HalvingGridSearchCV(clf, param_grid, cv = cv, random_state=42, scoring = 'f1', verbose=3, refit=True).fit(X_res, y_res)
print(search.best_params_)
print(search.best_score_) 
```

### Utilizes HalvingGridSearchCV to find the optimal hyperparameters for an SVC model, aiming to maximize the F1 score on the augmented dataset obtained after applying SMOTE. The best hyperparameters and the corresponding best F1 score are then printed for further analysis.

```python
clf = search.best_estimator_
from sklearn.model_selection import cross_val_predict
y_pred = cross_val_predict(clf, X_res, y_res, cv=cv)
print(classification_report(y_res, y_pred, target_names=['healthy', 'pathological']))
```

### Evaluates the performance of the best estimator using cross-validation and provides a detailed report on how well the model performs for each class (healthy and pathological), including metrics such as precision, recall, and F1-score.

```python
ConfusionMatrixDisplay.from_predictions(y_res, y_pred, cmap='coolwarm', normalize = 'true')
plt.show()
```

Computes the confusion matrix using the true labels (`y_res`) and predicted labels (`y_pred`), and then visualizes it using `ConfusionMatrixDisplay`. The `display_labels` parameter specifies the labels to be displayed on the plot.
