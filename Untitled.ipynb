### VaR:
VaR is one way professional traders manage risk. It estimates how much your portfolio might lose over a set time period. Using VaR is a good way to avoid losing all your money if the market moves against you.

```python
%matplotlib inline

import pandas as pd
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
import datetime as dt

import yfinance as yf
```

Simulate portfolio of stocks

```python
# create your portfolio of stocks
tickers = ["AAPL", "TSLA", "C", "DIS"]

# pick the weights of each stock (must add to 1)
weights = np.array([0.25, 0.3, 0.15, 0.3])

# the size of the portfolio
portfolio_value = 1_000

# the confidence interval (95%)
confidence = 0.05
```

The stock data for all the symbols in one line of code.

```python
data = yf.download(tickers, start="2020-01-01", end="2023-12-31")["Close"]
```

### Portfolio Statistics

Computing portfolio returns is not as simple as just adding up the returns of the individual stocks. We need to take the covariance between the stocks in the portfolio into account.

```python
# compute daily returns of each stock
returns = data.pct_change()

# compute the daily mean returns of each stock
mean_returns = returns.mean()

# compute portfolio mean return
port_mean = mean_returns.dot(weights)

# mean of investment returns
investment_mean = (1 + port_mean) * portfolio_value

# compute the portfolio covariance matrix
cov_matrix = returns.cov()

# compute portfolio standard deviation
port_stdev = np.sqrt(weights.T.dot(cov_matrix).dot(weights))

# standard deviation of investment returns
investment_stdev = portfolio_value * port_stdev
```

First, we get the daily returns of the stocks in the portfolio. From there we get the mean return for all the data. We apply the weights to those returns and multiply them by the portfolio value to get the portfolio mean return.

Then we compute the covariance between the returns, take the square root of the covariance-adjusted weights of the stocks in the portfolio, and compute the portfolio standard deviation.

The portfolio mean and standard deviation are used in the next step.

### Compute VaR

To find the VaR of this portfolio, we start by finding the point on the density plot based on the confidence level, mean, and standard deviation.

```python
# ppf takes a percentage and returns a standard deviation 
# multiplier for what value that percentage occurs at.
# It is equivalent to a one-tail test on the density plot.
percent_point = norm.ppf(confidence, investment_mean, investment_stdev)
```

VaR is the portfolio value less this amount.



```python
# calculate the VaR at the confidence interval
value_at_risk = portfolio_value - percent_point

# print it out
f"Portfolio VaR: {value_at_risk}"
```

This is the most you can expect to lose in one day with 95% confidence.


Stock returns increase with the square root of time. Therefore to scale the value at risk, we multiply it by the square root of time.

Multiply the one-day VaR by the square root of one (day) through the square root of 30 (days).

```python
value_at_risks = value_at_risk * np.sqrt(range(1, 31))
```

### PLOT

```python
# build plot of VaR over time
plt.xlabel("Day")
plt.ylabel("Max loss")
plt.title("Portfolio VaR")
plt.plot(value_at_risks, "r")
```

### PORTFOLIO OPTIMIZATION with skfolio:

Portfolio optimization is a balance between maximizing returns and minimizing risk.

First, we have to accurately forecast future returns and risk.

Then, we have to use tricky optimization models to build the portfolios subject to our constraints.

Over time, methods have been introduced to address the issues of mean-variance optimization. These include covariance shrinkage, regularization, using different risk measures, among others.

"skfolio" addresses the sprawling list of techniques used to optimize portfolios. It’s a library for portfolio optimization built on top of scikit-learn. It lets us easily build, fine-tune, and cross-validate portfolio models

benchmark: the benchmark for the backtesting. SPY is the default benchmark.

rebalance: rebalance frequency (for calendar rebalancing) or custom rebalance dates (in a list, format: YYYY/MM/DD)

max_weights: maximum proportion of capital you can invest in a single asset.

risk_manager: risk management for your strategy (Stop Loss, Take Profit, Max Drawdown).
